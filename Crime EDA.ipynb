```python
data_sorted.groupby('Year').count()['DR_NO'].plot(kind='bar')
plt.title('Number of crimes commited from 2020 to November 2023')
plt.ylabel('# of crime')
```

```python
age_data = data_sorted['Vict Age'].dropna()
```

```python
bins = [1, 18, 25, 35, 45, 55, 65, 75, 100]
labels = ['1-18', '19-25', '26-35', '36-45', '46-55', '56-65', '66-75', '76+']
```

```python
data_sorted['Age Group'] = pd.cut(age_data, bins=bins, labels=labels, right=False)
```

```python
plt.figure(figsize=(10, 6))
data_sorted['Age Group'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title('Distribution of Crime Victims by Age Group in LA from 2020 -2023')
plt.xlabel('Age Group')
plt.ylabel('Number of Crimes')
plt.show()
```

```python
plt.figure(figsize=(15, 10))
filtered_data = data_sorted[data_sorted['Vict Sex'].isin(['M', 'F', 'X'])]
filtered_data.groupby(['Year', 'Vict Sex']).count()['DR_NO'].unstack().plot(kind='bar', color=['pink','skyblue','red'])

plt.title('Number of Crimes by Year and Victim Sex')
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.legend(title='Victim Sex', loc='upper right')
plt.show()
```

```python
crime_by_vict_desc = data_sorted['Vict Descent'].value_counts().head(6)
```

```python
plt.figure(figsize=(10, 8))
pie=crime_by_vict_desc.plot(kind='pie', autopct='', startangle=90, cmap='viridis_r')
plt.ylabel('')
labels = [f'{l}, {s:0.1f}%' for l, s in zip(crime_by_vict_desc.index, crime_by_vict_desc / crime_by_vict_desc.sum() * 100)]
plt.legend(pie.patches, labels, loc='center left', bbox_to_anchor=(1, 0.5), title='Percentage')
plt.title('Victim of Crime in LA by Ethinic Groups')
plt.show()
```

```python
years = data_sorted['Year']

plt.figure(figsize=(15, 10))

for year in years.unique():
    filtered_year_data = data_sorted[data_sorted['Year'] == year]
    crime_by_area = filtered_year_data.groupby(['AREA NAME']).count()['DR_NO']
    
    # Create subplots
    plt.subplot(2, 2, years.unique().tolist().index(year) + 1)
    crime_by_area.plot(kind='bar', color='skyblue')
    plt.title(f'Number of Crimes by Area in {year}')
    plt.xlabel('Area Name')
    plt.ylabel('Number of Crimes')

plt.tight_layout()
plt.show()
```

```python
staus_crime = data_sorted['Status Desc'].value_counts()
```

```python
plt.figure(figsize=(10, 8))
pie=staus_crime.plot(kind='pie', autopct='', startangle=90, cmap='viridis')
plt.ylabel('')
labels = [f'{l}, {s:0.1f}%' for l, s in zip(staus_crime.index,staus_crime / staus_crime.sum() * 100)]
plt.legend(pie.patches, labels, loc='center left', bbox_to_anchor=(1, 0.5), title='Percentage')
plt.title('crime status for crimes in LA from 2020 -2023')
plt.show()
```
